### How could us treat the null numbers?

**Competition_distance: distance in meteres to the nearest competitor store**

Assumption: A null value exists when distance is so big that you don't need write
therefore can be used a high value to represent that

```python
df1.competition_distance = np.where(
   df1.competition_distance.isnull(),
   df1.competition_distance.max(),
   df1.competition_distance
   )

# another option is df1.competition_distance.fillna(200000)
```

**Competition_open_since_month and ..._year: approximate year and month of the time the nearest competitor was opened**

we start from the following assumption -> we cannot put this information as 0 because it is unlogical, because it is not possible for you to record data from a competition that has not opened the store (open_since = 0) so we will replace the null values of it with the same value of the column "date" we will consider this a presupposition with risk of fragility but we will follow with it from the CRISP method to test the model including it due to its importance, until the method proves its validity

```python
df1.competition_open_since_month.fillna(df1.date.dt.month, inplace=True)

df1.competition_open_since_year.fillna(df1.date.dt.year, inplace=True)
```

**Promo2_since_week: Describes the year and calendar week when the store started participating in Promo2**
This assumption to fill null values is much simple,
we'll substitute for "date" because the store will not participate in Promo2

```python
df1.promo2_since_week.fillna(df1.date.dt.isocalendar().week, inplace=True)

df1.promo2_since_year.fillna(df1.date.dt.isocalendar().year, inplace=True)
```

**Promo_interval: describes the conective intervals Promo2 is started**


```python
from datetime import datetime as dt

## creating a day_of_month_name dictionary to rename the column 'month_map'
day_of_month_name = {each : dt(month=each, year=1, day=1).strftime('%b') for each in range(1, 13)}
day_of_month_name
df1['month_map'] = df1.date.dt.month.replace(day_of_month_name)

## verify if month_map
df1.promo_interval.fillna('', inplace=True)
func = np.vectorize(lambda x, y: 1 if x in y.split(',') else 0)
df1['is_promo'] = func(df1['month_map'], df1['promo_interval'])
```

## 2.3. Feature Engineering:

```python
# year
df2.loc[:, 'year'] = df2.date.dt.isocalendar().year
# month
df2.loc[:,'month'] = df2.date.dt.month
# day
df2.loc[:, 'day'] = df2.date.dt.isocalendar().day
# week of year
df2.loc[:, 'week_of_year'] = df2.date.dt.weekofyear
# year week
df2.loc[:, 'year_week'] = df2.date.dt.strftime('%Y-%W')

# competition since
step = '01 '+ df2.competition_open_since_month.astype(str) + ' ' + df2.competition_open_since_year.astype(str)
df2.loc[:, 'competition_since'] = pd.to_datetime(step, format='%d %m %Y')
df2.loc[:, 'competition_time_month'] = (((df2['date'] - df2['competition_since'])/30).dt.days)

# promo since
df2.loc[:, 'promo_since'] = df2['promo2_since_year'].astype(str) + '-' + df2['promo2_since_week'].astype(str)
step = df2.loc[:, 'promo2_since_year'].astype(str) + '-' + df2['promo2_since_week'].astype(str)
df2.loc[:, 'promo_since'] = pd.to_datetime(step + '-1', format='%Y-%W-%w') - timedelta(days=7)
df2.loc[:, 'promo_time_week'] = ((df2['date'] - df2['promo_since'])/7).dt.days

# assortment
## a = basic, b = extra, c = extended
conditions = [df2['assortment'] == 'a', df2['assortment'] == 'b', df2['assortment'] == 'c']
results = 'basic extra extended'.split()
df2.loc[:, 'assortment'] = np.select(conditions, results)

# state holiday
results = 'public_holiday easter_holiday christmas'.split()
df2.loc[:, 'state_holiday'] = np.select(conditions, results)


del step, conditions, results
```
